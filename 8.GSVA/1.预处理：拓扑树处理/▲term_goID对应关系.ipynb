### 读取拓扑树

```python
# 读取树
GO_terms = pd.read_csv('▲GB_termTree.txt',sep = '\t',header=None,index_col=None)
GO_terms.columns = ['ID','Term_name','JSON']
```

```python
GO_terms
```

```python
# import json   
# sss = json.loads(GO_terms.iloc[1,2])    # json.loads() 加载json对象
# sss
```

```python
# 把GO_terms处理一下
# 1.小写转换成大写
# 2.“-”和“'”和“/”转换成空格
# 3.其他规则
for i in range(len(GO_terms.index)):
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].upper()
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].replace('(','_')
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].replace(')','_')
    
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].replace(',','_')
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].replace('-','_')
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].replace("'",'')
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].replace('/','_')
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].replace('.','_')
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].replace(':','_')
    
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].replace(' ','_')
    GO_terms['Term_name'][i] = GO_terms['Term_name'][i].replace('__','_')
    
    if GO_terms['Term_name'][i][-1]=='_':
        GO_terms['Term_name'][i] = GO_terms['Term_name'][i][:-1]
```

```python
GO_terms.set_index('Term_name', inplace = True)
```

```python
termsClass = pd.DataFrame(data={'gsva_name':terms2Bclass})  
termsClass.loc[:,'class_list'] = '[]'
termsClass
```

```python
nameList = GO_terms.index.tolist()
orphan = []

for terms in terms2Bclass:
    if terms not in nameList:
        orphan.append(terms)

print(f'{len(terms2Bclass)}个里面有{str(len(orphan))}个在拓扑树里没找到')
```
