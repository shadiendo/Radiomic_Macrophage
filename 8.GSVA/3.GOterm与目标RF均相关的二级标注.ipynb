# 根据P值筛选term

```python
# 对设定的满足个数
allNum = len(p_ret.columns)

howMuchCol = list(range(allNum+1))[::-1][:-3]

# howMuchCol = [allNum,allNum-1,allNum-2,allNum-3,allNum-4,allNum-5,allNum-6]
# howMuchCol = [allNum-3,allNum-4,allNum-5,allNum-6,allNum-7,allNum-8,allNum-9]
# howMuchCol = [allNum-5,allNum-6,allNum-7,allNum-8,allNum-9,allNum-10,allNum-11]
# howMuchCol = [allNum-6,allNum-7,allNum-8,allNum-9,allNum-10,allNum-11,allNum-12]
# howMuchCol = [allNum-7,allNum-8,allNum-9,allNum-10,allNum-11,allNum-12,allNum-13]

empty = [[] for i in range(len(howMuchCol))]
DDD = dict(zip(howMuchCol, empty))
df_DDD = dict(zip(howMuchCol, empty))
ddd_topMeaningfulTerms =dict(zip(howMuchCol, empty))

print(howMuchCol)

for num2satisfy in list(DDD.keys()):
    # 对每行验证
    for term in p_ret.index:
        # 对每行的每个数值验证
        corrNum = 0
        for i in p_ret.loc[term,:]:
            if i<=p_threshold:
                corrNum+=1
        
        if corrNum==num2satisfy:
            DDD[num2satisfy].append(term)

for key,values in DDD.items():
    print(f'{key}有{len(values)}个term满足')
    
for key,values in DDD.items():
    df = p_ret.T[values].T
    new_col = []
    haveImmuneProcess = []
    for row in df.index:
        # 添加一列，表示这个go条目属于哪些二级分类中
        whitchClass = GO_terms.loc[row,'class2level']
        new_col.append(whitchClass)
        # 添加一列，表示这些二级分类中是否含有 免疫过程
        if 'GO:0002376' in whitchClass:
            haveImmuneProcess.append('Yes')
        else:
            haveImmuneProcess.append('no')
        
    df.insert(0,'witchClass',new_col)
    df.insert(1,'haveImmuneProcess',haveImmuneProcess)
    
    # 以是否含有免疫过程这列作为条件排序
    df.sort_values("haveImmuneProcess",inplace=True)

    df_DDD[key] =df

# ##################################################
# 判断是否含有ImmuneProcess
for key,df in df_DDD.items():
    haveImmuneProcess = df['haveImmuneProcess'].tolist()
    if 'Yes' in haveImmuneProcess:
        print(f'当个数为{key}时，出现了目标，即immune system process')
        break
    print(f'当个数为{key}时，不存在immune system process')    
```

```python
# 删除所有 witchClass 为'not_find'的行
dictDf_cleaned =dict(zip(howMuchCol, empty))

for i in howMuchCol:
    dict_df = df_DDD[i]
    dict_df = dict_df.drop(dict_df[(dict_df['witchClass']=='not_find')].index)
    dictDf_cleaned[i]=dict_df
```

```python
# 找出列表中比 阈值【大】的 元素索引
def findListIndex_biggerThanThredhold(list_demo,thredhold_demo):
    ret = []
    for i in range(len(list_demo)):
        if list_demo[i] > thredhold_demo:
            ret.append(i)
    return ret

# demolist=[0.6490543637619276, 0.017827062594293376, 2.0362808329794554e-05, 5.276562604509389e-07, 0.29530070116013163, 0.3947785711498213, 0.13187170281116153, 0.07259014107606869, 0.00038957803770905215, 0.00020408525052975152, 0.0003275962143213668, 0.003415254709055932, 0.01115993083762145, 0.14135822169622228]
# findListIndex_biggerThanThredhold(demolist,p_threshold)
```

```python
df_retCoor = GoTermClassRet.copy(deep=True)
for row_id in range(len(df_retCoor.index)):
    for col_id in range(len(df_retCoor.columns)-1):
        col_id = col_id+1
        # 所有要替换的单元格
        cell = df_retCoor.iloc[row_id,col_id]
        print(f'当前单元格为{cell}')
        
        cell_value = []
        
        if cell=='':
            cell_value = 0
        else:
            for i in eval(cell):
                print(f'【{i}】')
                # 找对应的行
                aim_row_p = p_ret.loc[i,].tolist()
                aim_row_corr = corr_ret.loc[i,].tolist()
                # print(f'【P】{aim_row_p}\n')
                # print(f'【Coor】{aim_row_corr}\n')
                
                # 找到比阈值大的索引
                deleteIndex = findListIndex_biggerThanThredhold(aim_row_p,p_threshold)
                aim_row_corr  = [n for i, n in enumerate(aim_row_corr) if i not in deleteIndex]
                # print(f'根据p值确定的删除索引为{deleteIndex}')
                # print(f'【Coor】{aim_row_corr}\n')
                
                # ###############
                # 求个平均
                # ###############
                term_corr_mean = sum(aim_row_corr) / len(aim_row_corr)
                print(f'该term的平均相关度为{term_corr_mean}')
                
                cell_value.append(term_corr_mean)
        
        if cell_value!=0:
            cell_value = sum(cell_value) / len(cell_value)
        print(f'当前单元格的平均相关度为{cell_value}\n')
            
        df_retCoor.iloc[row_id,col_id] = cell_value
```

# 最终处理成能被R做成气泡热图的表
写个函数，把平铺开来的表，变成竖着的

```python
"""
df_demo:输入的df
colToHandle_demo:输入的要被处理的列
valueName_demo:处理后的列名
"""
def DfChangeDimention(df_demo,colToHandle_demo,valueName_demo):
    # 获取要保留的列
    reserveCol = list(set(df_demo.columns.tolist())-set(colToHandle_demo))
    
    # 逐个开始，获取df

    df_list = []
    for i in colToHandle_demo:
        df_demo_reserveCol = df_demo[reserveCol]  # 原来的列
        list_singleCol = df_demo[i].values.tolist()  # 新列
        
        # 在只包含原来的列df中加入新的单独一列
        df_demo_reserveCol.insert(len(reserveCol), valueName_demo, list_singleCol)
        # 把分类名重复n次之后加入df中
        class_col = [i for j in range(df_demo.shape[0])]
        df_demo_reserveCol.insert(len(reserveCol), 'class', class_col)
        
        # 把这个df加入列表中
        df_list.append(df_demo_reserveCol)
        
    # 把上面的拼起来
    df_ret = pd.concat(df_list,axis =0,ignore_index=False)
    
    return df_ret
```

```python
# 将 Count 表 胖子变瘦子
columnToProcess = df_retCount.columns.tolist()[1:]
df_count_FAT  = DfChangeDimention(df_retCount,columnToProcess,'Count')

# 将 Count_log 表 胖子变瘦子
columnToProcess = df_retCount_log.columns.tolist()[1:]
df_count_log_FAT  = DfChangeDimention(df_retCount_log,columnToProcess,'Count_log')

# 将 coor 表 胖子变瘦子
columnToProcess = df_retCoor.columns.tolist()[1:]
df_coor_FAT  = DfChangeDimention(df_retCoor,columnToProcess,'coor')
```
