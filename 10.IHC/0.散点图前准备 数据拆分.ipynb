```python
import pandas as pd
import numpy as np
```

```python
# ━━━━━━━━━━━━━━━━━━━━━━━━  打开影像组学文件
df_radiomics = pd.read_table('→RiskScore_影像组学数据.csv',sep=',',index_col=0)

# 新增一列,分组
cut_off = 2
which_BOX_list = []
for row in df_radiomics.index:
    RiskScore_h0 = df_radiomics.loc[row,'RiskScore_h0']
    if RiskScore_h0<cut_off:
        which_BOX_list.append('low')
    else:
        which_BOX_list.append('high')
    
df_radiomics.insert(1, 'WhichRiskBox_h0', which_BOX_list)  # 新列


df_radiomics = df_radiomics.T[:2].T
df_radiomics
```

```python
# ━━━━━━━━━━━━━━━━━━━━━━━━  打开IHC结果文件
df_IHC = pd.read_table(r'蜡块数据_IHC-profilered.csv',sep=',',index_col=0)
df_IHC = df_IHC.T[1:].T
df_IHC
```

```python
# ━━━━━━━━━━━━━━━━━━━━━━━━  获取交叉
df_IHC_list = df_IHC.index.tolist()
df_radiomics_list = df_radiomics.index.tolist()
patCrossCondition = list(set(df_radiomics_list) & set(df_IHC_list))
print(f'有{len(patCrossCondition)}个交叉')

df_tmp = pd.concat([df_radiomics,df_IHC],axis=1).T[patCrossCondition].T
df_tmp
```

```python
df_tmp.to_csv('→→IHC和影像组学数据交集.csv',index=True,header=True)
```

# 重构

```python
# 要把 数据列 变成分类的 特征名们
features_all = df_tmp.columns.tolist()
features = features_all[2:]
print(f'【{df_tmp.shape[0]}行数据】*【{len(features)}个选中特征】=【{df_tmp.shape[0]*(len(features))}行数据】')

features
```

```python
df_list = []
reserveList=[i for i in features_all if i not in features]

for i in features:
    df_single = df_tmp[reserveList]  # 原来的列
    new_col = df_tmp[i].values.tolist()  # 新列
    
    # 在第一列加入新的
    df_single.insert(0, 'enrichment_value', new_col)
    # 把分类重复n次之后加入df中
    class_col = [i for j in range(df_tmp.shape[0])]
    df_single.insert(0, 'class', class_col)
    
    df_list.append(df_single)
df_list[0]
```

```python
# 把散装的合并起来
df_ret = pd.concat(df_list,axis =0,ignore_index=False)
df_ret
```

```python
df_ret.to_csv('→→长数据_供绘散点图.csv',index=True,header=True)
```
