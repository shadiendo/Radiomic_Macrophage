```python
# 主程序
def conbineCsvFromFilieList(csv_file_list,save_name):
    csv_list = []
    for csv in csv_file_list:
        df = pd.read_csv(csv,index_col=0).set_index(keys='featureName')
        csv_list.append(df)
        
    csv_conbine = pd.concat(csv_list,axis=1)
    csv_conbine.to_csv(save_name, index=True)
    print(f'已成功保存至',save_name)
    
    return csv_conbine
```
